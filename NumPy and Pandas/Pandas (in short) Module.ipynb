```python
# list of dictionary 
weather_data = [    
    {'day': '1/1/2017', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},
    {'day': '1/2/2017', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'},
    {'day': '1/3/2017', 'temperature': 28, 'windspeed': 2, 'event': 'Snow'},
]

print(type(weather_data))  #Printing data type of weather_data

df1 = pd.DataFrame(data=weather_data, columns=['day','temperature','windspeed','event'])    #Convert it into DataFrame

print(type(df1))
df1    #printing dataframe
```
